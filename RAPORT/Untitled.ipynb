```python
import os.path
import subprocess
import pandas as pd
import zipfile

import matplotlib.pyplot as plt

dataset_name = "oles04/top-leagues-player"


subprocess.run(["kaggle", "datasets", "download", "-d", dataset_name])

# Directory 
ls = "/Users/maxon462/Desktop/PWR/Metody systemowe i decyzyjne/RAPORT/pythonProject"

# Check if the ZIP file exists before extracting
zip_file = os.path.join(ls, 'top-leagues-player.zip')
if os.path.exists(zip_file):
    # Extract the ZIP file
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(ls)

    
    csv_file = "/Users/maxon462/Desktop/PWR/Metody systemowe i decyzyjne/RAPORT/pythonProject/top5_leagues_player.csv"
    df = pd.read_csv(csv_file)
    print("Dataset loaded successfully.")
else:
    print("ZIP file not found.")
```

```python

df = df.drop(columns=['full_name', 'place_of_birth', 'player_agent'])

# Extract the position category(Only defender, mid and attack)
df['position'] = df['position'].str.split('-').str[0].str.strip()

df['nationality'] = df['nationality'].str.split().str[0]






df.head(50)
```

```python
position_counts = df['position'].value_counts()

# Step 3: Plot the counts using matplotlib
position_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Players by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.tight_layout()  
plt.show()
```

```python

nationality_counts = df['nationality'].value_counts().head(10)  
nationality_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Nationalities of Players')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()

plt.figure(figsize=(8, 6))  # Set figure size
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')  
plt.title('Distribution of Player Ages', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()

```

```python
average_height_by_position = df.groupby('position')['height'].mean().sort_values()


plt.figure(figsize=(10, 6))  # Larger figure size
plt.bar(average_height_by_position.index, average_height_by_position.values, color='lightgreen', width=0.5)  
plt.title('Average Height of Players by Position', fontsize=16)  
plt.xlabel('Position', fontsize=14)  
plt.ylabel('Average Height (cm)', fontsize=14)  
plt.grid(axis='y', linestyle='--', alpha=0.7)   
plt.tight_layout()  
plt.show()
```

```python
# plt.figure(figsize=(8, 6))  # Set figure size
plt.scatter(df['age'], df['price'], color='skyblue', alpha=0.7)  
plt.title('Player Age vs. Price', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Price in mln', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```
